### 1. Improved readability  
A CTE makes it easier to interpret a query compared to using a subquery. 
  
Finding employees from the Chicago office who do not have the salesperson title.

Fulfilling the query without a CTE.

````python
%%sql
SELECT *
FROM employee
WHERE office = 'Chicago'
    AND title != 'salesperson'
ORDER BY title;
````

Using a CTE to find all employees from the Chicago office. Querying from the CTE then filtering for employees without the salesperson title. Sorting the results by the title.

````python
%%sql
WITH employee_chicago AS (
    SELECT *
    FROM employee
    WHERE office = 'Chicago'
)
SELECT *
FROM employee_chicago
WHERE title != 'salesperson'
ORDER BY title;
````

Replacing the CTE with a subquery to create a derived table.

````python
%%sql
SELECT *
FROM
(
    SELECT *
    FROM employee
    WHERE office = 'Chicago'
) AS employee_chicago
WHERE title != 'salesperson'
ORDER BY title;
````

### 2. Nesting  
We can refer to other CTEs within a CTE. Managing the complexity added when using nested subqueries to create derived tables.  

```
WITH cte1 AS (
    SELECT col1, col2
    FROM table1
),
cte2 AS (
    SELECT col1, col2
    FROM cte1
)
SELECT cte1.col2 
FROM cte1 
JOIN cte2
    ON cte1.col1 = cte2.col1;
```

Finding the employees from the Chicago office who have dba as their title. Using 2 separate CTEs
1. finding employees from the Chicago office 
2. filtering the Chicago employees to only show those who have dba as their title.

````python
%%sql
WITH employee_chicago AS (
    SELECT *
    FROM employee
    WHERE office = 'Chicago'
),
employee_chicago_dba AS (
    SELECT *
    FROM employee_chicago
    WHERE title = 'dba'
)
SELECT *
FROM employee_chicago_dba
````

#### 3. Using CTEs for year-over-year comparisons  
CTE query can be used multiple times in the same query.  
  
Comparing the annual commission totals for the salespersons from one year to the next. 

````python
%%sql
SELECT 
    salesperson_id,
    YEAR(commission_date) AS year,
    SUM(commission_amount) AS commission_total
FROM commission
GROUP BY 
    salesperson_id,
    year
ORDER BY salesperson_id;
````

Displaying the current and previous year's commission totals on the same row and creating a CTE to run the aggregate query. Using SELECT from the CTE twice with a JOIN on salesperson_id to get the results for the current and previous year. Creating different table aliases for the same CTE. Adding a filter to return where the current year equals the last year plus 1.

````python
%%sql
WITH commission_year AS(
    SELECT 
    salesperson_id,
    YEAR(commission_date) AS year,
    SUM(commission_amount) AS commission_total
FROM commission
GROUP BY 
    salesperson_id,
    year
)
SELECT *
FROM commission_year current_year
JOIN commission_year previous_year
    ON current_year.salesperson_id = previous_year.salesperson_id
WHERE current_year.year = previous_year.year + 1
ORDER BY current_year.salesperson_id;
````

Modifying the query above to display the salesperson's name.

````python
%%sql
WITH commission_year AS(
    SELECT
    salesperson_id,
    employee.name,
    YEAR(commission_date) AS year,
    SUM(commission_amount) AS commission_total
FROM commission
JOIN employee
    ON commission.salesperson_id = employee.employee_id
GROUP BY 
    salesperson_id,
    year
)
SELECT 
    current_year.salesperson_id AS current_year_saleperson_id,
    current_year.name AS salesperson_name,
    current_year.year AS current_year,
    FORMAT(current_year.commission_total,2) AS current_year_commission,
    previous_year.year AS previous_year,
    FORMAT(previous_year.commission_total,2) AS previous_year_commission
FROM commission_year current_year
JOIN commission_year previous_year
    ON current_year.salesperson_id = previous_year.salesperson_id
WHERE current_year.year = previous_year.year + 1
ORDER BY current_year.salesperson_id;
````

Add the commission total difference between the two years into the SELECT. Sort the results by that difference.

````python
%%sql
WITH commission_year AS(
    SELECT
    salesperson_id,
    employee.name,
    YEAR(commission_date) AS year,
    SUM(commission_amount) AS commission_total
FROM commission
JOIN employee
    ON commission.salesperson_id = employee.employee_id
GROUP BY 
    salesperson_id,
    year
)
SELECT 
    current_year.salesperson_id AS current_year_saleperson_id,
    current_year.name AS salesperson_name,
    current_year.year AS current_year,
    FORMAT(current_year.commission_total,2) AS current_year_commission,
    previous_year.year AS previous_year,
    FORMAT(previous_year.commission_total,2) AS previous_year_commission,
    FORMAT(current_year.commission_total - previous_year.commission_total,2) AS commission_difference
FROM commission_year current_year
JOIN commission_year previous_year
    ON current_year.salesperson_id = previous_year.salesperson_id
WHERE current_year.year = previous_year.year + 1
ORDER BY current_year.commission_total - previous_year.commission_total;
````

Modifying the output to only show the salespeople whose sales went down from year to year.

````python
%%sql
WITH commission_year AS(
    SELECT
    salesperson_id,
    employee.name,
    YEAR(commission_date) AS year,
    SUM(commission_amount) AS commission_total
FROM commission
JOIN employee
    ON commission.salesperson_id = employee.employee_id
GROUP BY 
    salesperson_id,
    year
)
SELECT 
    current_year.salesperson_id AS current_year_saleperson_id,
    current_year.name AS salesperson_name,
    current_year.year AS current_year,
    FORMAT(current_year.commission_total,2) AS current_year_commission,
    previous_year.year AS previous_year,
    FORMAT(previous_year.commission_total,2) AS previous_year_commission,
    FORMAT(current_year.commission_total - previous_year.commission_total,2) AS commission_difference
FROM commission_year current_year
JOIN commission_year previous_year
    ON current_year.salesperson_id = previous_year.salesperson_id
WHERE current_year.year = previous_year.year + 1
    AND current_year.commission_total < previous_year.commission_total
ORDER BY current_year.commission_total - previous_year.commission_total;
````

### 4. Comparing individuals against their own group  

Using a CTE in the WHERE.  
  

Finding all the salespeople who made at least 2% of the total commissions earned by all salespeople in the company for 2017. Reusing the query above to get the commission totals and adding it to a CTE but only including data for 2017. Calculating the 2 percent of the total commissions earned in the WHERE clause by querying the CTE a second time.

Filter the commision totals query to only include 2017.

````python
%%sql
    SELECT
    salesperson_id,
    employee.name,
    SUM(commission_amount) AS commission_total
FROM commission
JOIN employee
    ON commission.salesperson_id = employee.employee_id
WHERE commission_date BETWEEN '2017-01-01' AND '2017-12-31 23:59:59'
GROUP BY 
    salesperson_id

````

What is 2% of the commission total?

````python
%%sql
WITH commission_year AS (
    SELECT
        salesperson_id,
        employee.name,
        SUM(commission_amount) AS commission_total
    FROM commission
    JOIN employee
        ON commission.salesperson_id = employee.employee_id
    WHERE commission_date BETWEEN '2017-01-01' AND '2017-12-31 23:59:59'
    GROUP BY 
        salesperson_id
)
SELECT
    SUM(commission_total)*.02
FROM commission_year;
````

Finding all the salespeople who made at least 2% of the total commissions earned by all salespeople in the company for 2017.

````python
%%sql
WITH commission_year AS (
    SELECT
        salesperson_id,
        employee.name,
        SUM(commission_amount) AS commission_total
    FROM commission
    JOIN employee
        ON commission.salesperson_id = employee.employee_id
    WHERE commission_date BETWEEN '2017-01-01' AND '2017-12-31 23:59:59'
    GROUP BY 
        salesperson_id
)
SELECT *
FROM commission_year cyl
WHERE commission_total >
    (
        SELECT
            SUM(commission_total)*.02
        FROM commission_year
    )
ORDER BY commission_total DESC;
````
